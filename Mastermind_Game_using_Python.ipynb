```python
import random  # Used to generate a random number

# Generate a random 4-digit number between 1000 and 9999
num = random.randint(1000, 9999)

# Ask the user for their first guess
n = int(input("Guess the 4 digit number: "))

# Check if the user guessed correctly on the first try
if n == num:
    print("Great! You guessed the number in just 1 try. You are a Mastermind!")

else:
    ctr = 1  # Counts how many guesses the user has made

    # Keep asking until the user guesses the correct number
    while n != num:
        ctr += 1  # Increase attempt counter

        # Convert numbers to strings so we can compare digit by digit
        n_str = str(n)
        num_str = str(num)

        count = 0              # Number of matching digits in the correct place
        correct = ['X'] * 4    # Stores correct digits found by the user

        # Compare each of the 4 positions
        for i in range(4):
            if n_str[i] == num_str[i]:
                count += 1            # Increase match count
                correct[i] = n_str[i] # Store the correct digit at the correct spot

        # Give feedback based on how many digits matched
        if count > 0:
            print("Not quite the number. But you did get", count, "digit(s) correct!")
            print("Correct digits in correct positions:", correct)
        else:
            print("None of the digits match the actual number.")

        print()
        # Ask the user for another guess
        n = int(input("Enter your next 4-digit guess: "))

    # This runs after the loop ends (meaning user guessed correctly)
    print("You've become a Mastermind!")
    print("It took you only", ctr, "tries.")
```
